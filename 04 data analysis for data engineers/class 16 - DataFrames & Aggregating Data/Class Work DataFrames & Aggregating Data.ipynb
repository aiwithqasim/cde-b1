### 2- Creating New Columns

```python
titanic['Discounted_Fare'] = titanic['Fare']/10
titanic
```

```python
# if age < 20  20% discount
# 20 - 30 15 % discount
# 30+ 5 % discount
```

```python
import numpy as np

titanic['Discounted Fares'] = np.where(titanic['Age']<20,titanic['Fare']%20,titanic['Fare'])
```

```python
titanic['Discounted Fares']
```

```python
def apply_discount(age):
    if age<20:
        return 0.2
    elif 20<= age <30:
        return 0.15
    elif age>30:
        return 0.5

titanic['Discount']=titanic['Age'].apply(apply_discount)

titanic['Dicounted_Fare'] = titanic['Fare']*(1-titanic['Discount'])

print(titanic[['Age','Discount','Fare','Discounted_Fare']])
```

### .agg() method

```python
def pct_30(column):
    return column.quantile(0.3)

titanic['Age'].agg(pct_30)
# Single Operation
```

```python
df = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, np.nan]],
                  columns=['A', 'B', 'C'])
df
```

```python
def pct_30(column):
    return column.quantile(0.3)
def pct_40(column):
    return column.quantile(0.4)


# titanic['Age'].agg([pct_30,pct_40])
titanic['Age'].agg(lambda x: x.quantile(0.3))
```

### Pivot tables


```python
#pivot and implicitly define agffunc=np.mean
titanic.pivot_table(values = 'Age', index='Sex')
```

```python
#explicitly define statistics i:e np.median
titanic.pivot_table(values= 'Age', index='Sex', aggfunc=np.median)
```

```python
#multiple statistics
titanic.pivot_table(values='Age', index='Sex', aggfunc=[np.std, np.median,np.mean])
```

```python
#pivot on two varibales
titanic.pivot_table(values='Age', index='Sex', columns='Survived')
# aggfunc = np.mean
```

```python
titanic.pivot_table(values='Age', index='Sex', columns='Survived', fill_value=0)
```
